# Join Datasets together

<img src="brazilian-ecommerce/join_schema.png">

```python
#joining category english name to products table
products = pd.merge(products, category, how='left', left_on=['product_category_name'],
                           right_on=['product_category_name'])
```

```python
len(products)
```

```python
products.isnull().sum()
```

```python
len(customers)
```

```python
len(sellers)
```

```python
#join customers to order dataset
orders_customers = pd.merge(orders, customers, how='left', left_on=['customer_id'],
                           right_on=['customer_id'])
```

```python
#joins sellers to items
item_sellers=pd.merge(order_items,sellers,  how='left', left_on=['seller_id'],
                           right_on=['seller_id'])
```

```python
#join products to above dataset
orders_products = pd.merge(item_sellers, products, how='left', left_on=['product_id'],
                           right_on=['product_id'])
```

```python
len(orders_customers)
```

```python
len(orders_products)
```

```python
#join order payments to order dataset
orders_payments = pd.merge(orders_customers, order_payments, how='left', left_on=['order_id'],
                                                      right_on=['order_id'])
```

```python
#join order payments to order dataset
final = pd.merge(order_payments, orders_products, how='left', left_on=['order_id'],
                                                      right_on=['order_id'])
```

```python
len(final)
```

```python
del category
del customers
del orders
del sellers
del order_items
del products
del item_sellers
del order_payments
del orders_customers
del reviews
```
